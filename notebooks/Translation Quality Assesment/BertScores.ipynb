```python
import pandas as pd
from transformers import BertTokenizer, BertModel
from bert_score import BERTScorer
```

```python
df=pd.read_csv("transquest_scores.csv")
```

```python
df.columns
```

```python
def calculate_bertscore(df, col1, col2, model_type='bert-base-multilingual-cased'):
    """
    Calculate BERTScore for two columns of a DataFrame.

    Args:
        df (pandas.DataFrame): DataFrame containing the text columns.
        col1 (str): Name of the column with candidate translations.
        col2 (str): Name of the column with reference translations.
        model_type (str): Pretrained model to use for scoring (default: 'bert-base-multilingual-cased').

    Returns:
        dict: Dictionary with mean Precision, Recall, and F1 scores.
    """

    candidates = df[col1].tolist()
    references = df[col2].tolist()


    scorer = BERTScorer(model_type=model_type)


    P, R, F1 = scorer.score(candidates, references)


    return {
        "Precision": P.mean().item(),
        "Recall": R.mean().item(),
        "F1": F1.mean().item()
    }
```

```python
calculate_bertscore(df,"model_1_mr_text","model_2_mr_text")
```
